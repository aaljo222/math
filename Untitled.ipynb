```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


fig = plt.figure()
ax = fig.add_subplot(projection='3d')

colors = ['r', 'g', 'b', 'y']
yticks = [3, 2, 1, 0]
for c, k in zip(colors, yticks):
    # Generate the random data for the y=k 'layer'.
    xs = np.arange(20)
    ys = np.random.rand(20)

    # You can provide either a single color or an array with the same length as
    # xs and ys. To demonstrate this, we color the first bar of each set cyan.
    cs = [c] * len(xs)
    cs[0] = 'c'

    # Plot the bar graph given by xs and ys on the plane y=k with 80% opacity.
    ax.bar(xs, ys, zs=k, zdir='y', color=cs, alpha=0.8)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# On the y-axis let's only label the discrete values that we have data for.
ax.set_yticks(yticks)

plt.show()
```

$ \lim_{h\rightarrow 0 }  \frac{f(x+h) -f(x) }{ h} $

```python
import numpy as np
import matplotlib.pyplot as plt

# 함수 정의
def f(x):
    return 3*x**3 + 6*x

def derivative(x):
    return 9*x**2 + 6

# x 범위
x = np.linspace(-5, 5, 400)

# 원래 함수와 미분 함수
y = f(x)
y_prime = derivative(x)

# 그래프 그리기
plt.figure(figsize=(8,6))
plt.plot(x, y, label=r"$f(x) = 3x^3 + 6x$", color="blue")
plt.plot(x, y_prime, label=r"$f'(x) = 9x^2 + 6$", color="red", linestyle="--")

# 좌표축 꾸미기
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Function and Its Derivative")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 원래 함수 정의
def f(x):
    return 3*x**3 + 6*x

# 미분 정의를 활용한 수치 미분 (극한 근사)
def derivative_limit(x):
    h= 0.000001
    return (f(x+h) - f(x)) / h

# x 구간 설정
x = np.linspace(-5, 5, 400) # -5부터 5까지 간격을 400으로 쪼개라 

# 함수 값
y = f(x)

# 정의를 활용한 미분 근사
y_prime_approx = derivative_limit(x)

# 실제 해석적 미분 (검증용)
y_prime_exact = 9*x**2 + 6

# 그래프 그리기
plt.figure(figsize=(9,6))
plt.plot(x, y, label=r"$f(x) = 3x^3 + 6x$", color="blue")
plt.plot(x, y_prime_approx, label=r"$f'(x)$ (limit definition, approx)", color="red", linestyle="--")
plt.plot(x, y_prime_exact, label=r"$f'(x) = 9x^2 + 6$ (exact)", color="green", linestyle=":")

plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Derivative using Limit Definition")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 원래 함수 정의
def f(x):
    return 3*x**2 + 7

# 미분 정의를 활용한 수치 미분 (극한 근사)
def zz(x):
    h= 0.000001
    return (f(x+h) - f(x)) / h

# x 구간 설정
x = np.linspace(-5, 5, 400) # -5부터 5까지 간격을 400으로 쪼개라 

# 함수 값
y = f(x)

# 정의를 활용한 미분 근사
y_prime_approx = zz(x)

# 실제 해석적 미분 (검증용)
y_prime_exact = 6*x

# 그래프 그리기
plt.figure(figsize=(9,6))
plt.plot(x, y, label=r"$f(x) = 3x^3 + 6x$", color="blue")
# plt.plot(x, y_prime_approx, label=r"$f'(x)$ (limit definition, approx)", color="red", linestyle="--")
plt.plot(x, y_prime_exact, label=r"$f'(x) = 9x^2 + 6$ (exact)", color="green", linestyle=":")

plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Derivative using Limit Definition")
plt.legend()
plt.grid(True)
plt.show()
```
